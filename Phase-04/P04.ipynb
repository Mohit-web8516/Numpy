```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1 = np.array([[1, 2, 3],[ 4, 5, 6]])
array2 = np.random.rand(3,3)
array3 = np.zeros((4,4))
```

```python
np.save('array1.npy', array1)
np.save('array2.npy', array2)
np.save('array3.npy', array3)
```

```python
loaded_array1 = np.load('array1.npy')
print(loaded_array1)
```

## Image import 

```python
from PIL import Image
import os

print("Exists:", os.path.exists("numpy_logo.png"))
print("Size:", os.path.getsize("numpy_logo.png"))

with open("numpy_logo.png", "rb") as f:
    print("First 20 bytes:", f.read(20))
```

## Image size

```python
import os

size_bytes = os.path.getsize("numpy_logo.png")
size_kb = size_bytes / 1024

print(f"Size: {size_bytes} bytes")
print(f"Size: {size_kb:.2f} KB")
```

```python
import numpy as np
from PIL import Image

img = Image.open("numpy_logo.png")

img_array = np.array(img)

print(img_array.shape)
print(img_array.dtype)

np.save("numpy_logo.npy", img_array)

print("Saved successfully!")
```

```python
try:
    logo = np.load('numpy_logo.npy')
    #Display
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy logo")
    plt.grid(False)

    dark_logo = 1 - logo
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy logo")
    plt.grid(False)


except FileNotFoundError:
    print("Numpy logo file not found ")
```
